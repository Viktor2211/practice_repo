```python
# 1) Сумма цифр
# Напиши def sum_digits(n: int) -> int, которая возвращает сумму цифр числа n (n может быть отрицательным).
# sum_digits(123) -> 6
# sum_digits(-507) -> 12

def sum_digits(n:int):
    n_abs = abs(n)
    n_str = str(n_abs)
    result = 0
    for num in n_str:
        result += int(num)
    return result

sum_digits(-507)
```

```python
# 2) Уникальные элементы по порядку
# Напиши def unique_keep_order(lst: list[int]) -> list[int], которая оставляет только первые вхождения, сохраняя порядок.
# [1,2,1,3,2,4] -> [1,2,3,4]

def unique_keep_order(lst: list[int]) -> list[int]:
    result_lst = []
    for num in lst:
        if num not in result_lst:
            result_lst.append(num)
        else:
            continue
    return result_lst

unique_keep_order([1,2,5,3,2,4])
```
